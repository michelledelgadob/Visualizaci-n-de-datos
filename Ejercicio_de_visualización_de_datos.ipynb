```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/CODING_DOJO/CÓDIGO/WEEK3/Dark Web Price Index 2021 (1).xlsx")
df
```

```python
df.groupby("Category")["Avg. dark web Price (USD)"].mean()
```

```python
df.groupby("Category")["Avg. dark web Price (USD)"].mean().plot(kind="bar",cmap="Set2")
plt.title('Mean Price (USD) by Category in dark web')
plt.xlabel('Category')
plt.grid()
```

```python
df.groupby("Category")["Avg. dark web Price (USD)"].max().plot(kind="bar",cmap="Paired")
plt.title('Max Avg. Price (USD) by Category in dark web')
plt.xlabel('Category')
plt.grid()
```

```python
df.groupby("Category")["Avg. dark web Price (USD)"].min().plot(kind="bar",cmap="Set2")
plt.title('Min Avg. Price (USD) by Category in dark web')
plt.xlabel('Category')
plt.grid()
```
